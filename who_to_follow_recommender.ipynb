### Recommendation using Collaborative filtering with Nearest Neighbors

```python
# Checking if neighbouring nodes in the graph are following one another
def follow_each_other(graph, user):
    return set(graph.neighbors(user))
```

```python
# Function to check for a user's followers of followers
def followers_of_followers(graph, user):
    x=[]
    for each in graph.neighbors(user):
        for item in graph.neighbors(each):
            x.append(item)
    return set(x)
```

```python
# Function to select only common followers between two users
def common_followers(graph, user1, user2):
    x1 = follow_each_other(graph, user1)
    x2 = follow_each_other(graph, user2)
    return set(x1 & x2)
```

```python
# Converting number maps to sorted list
def number_map_to_sorted_list(map):
    map = sorted(map.items(), key = itemgetter(1), reverse=True)
    return map
```

```python
# Calculating recommendation score
def calc_score(graph, user, each):  
    score = 0
    common = common_followers(graph, user, each)
    for item in common:
        score = score + 1/(len(follow_each_other(graph, item)))
    return score
```

```python
# Function to check for a mapped network of common follower
def network_map(graph, user):
    network_scores = dict()
    for each in graph.nodes():
        if(each != user):
            score = calc_score(graph, user, each)
            network_scores[each] = score
    return network_scores
```

```python
# Recommending users based on its network and similar preference for following amother user
def who_to_follow(graph, user):
    recommendations = []
    d = network_map(graph,user)
    d = sorted(d.items(), key = itemgetter(1), reverse=True)
    for i in range(0,10):
        recommendations.append(d[i])
    return recommendations
```

```python
# Ensuring a clean list of recommendations
def return_pure_list(recommendations):
    pure_list = []
    for each in recommendations:
        pure_list.append(each[0])
    return pure_list
```

```python
# To display the recommended nearest neighbors from user's network to follow
def show(graph, user):
        nx.draw_networkx_nodes(graph, pos=nx.spring_layout(graph),
                           nodelist=return_pure_list(who_to_follow(graph, user)),
                           node_color='b',
                           node_size=500,
                       alpha=0.8)
        print("Recommendation from your network: " )
        print(who_to_follow(graph, user))
```
